# Transaction by Customer

```python
# import pandas as pd
# import numpy as np
# import matplotlib as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read .CSV files into notebook
cust = pd.read_csv('cleaned_customer_list.csv')
sales = pd.read_csv('transaction_data.csv')
```

```python
# merge and review data
cust_sales = pd.merge(cust, sales, left_on='cust_id', right_on='CustID', how='inner')
cust_sales
```

## Normalize Data

```python
# drop unnecessary columns
cust_sales = cust_sales.drop(columns=['Tax(8%)','Subtotal', 'UnitPrice', 'ProductVendor', 'VendorID', 'email', 'Employee'])
cust_sales
```

```python
# groupby custID avg
avg_sales = cust_sales.groupby(['name','CustID']).mean(['OrderTotal'])
avg_sales
```

```python
# drop unnecessary columns
avg_sales = avg_sales.drop(columns=['cust_id','TransactionID','ProductID'])
avg_sales
```

```python
# groupby custID sum
total_sales = cust_sales.groupby(['name','CustID']).sum(['OrderTotal'])
total_sales
```

```python
# drop unnecessary columns
total_sales = total_sales.drop(columns=['cust_id','TransactionID','ProductID'])
total_sales
```
